**exercicio 2.4 plotando dois graficos ao mesmo tempo**

```python
x2 = linspace(1, 10, 100)

y21 = sin(x2)
y22 = cos(x2)

plot(x2, y21, 'c--')
plot(x2, y22, 'k-')

show()
```
