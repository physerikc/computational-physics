**Problema 3.1 - plotando dados experimentais**

```python
def somat(data, r, i):
  soma = 0
  if i <= r:
    for m in range(-i, r+1):
      soma += data[i+m, 1]
  elif i >= len(data)-r:
    for m in range(-r, len(data)-i):
      soma += data[i+m, 1]
  else:
    for m in range(-r, r+1):
      soma += data[i+m, 1]
  return soma

data = np.loadtxt('sunspots.txt')

r = 5
Yk = np.zeros(len(data[:, 0]))

for i in range(0, len(Yk)):
  Yk[i] += (1/(2*r))*somat(data, r, i)

x = data[:,0]
y = data[:, 1]

plt.xlim(0, 1000)

plt.plot(x, y)
plt.plot(x, Yk)
plt.show()
```

**Problema 3.2 - plotando curvas**

```python
import numpy as np
import matplotlib.pyplot as plt

def theta_func(coeff_pi, nums):
    return np.linspace(0, coeff_pi*np.pi, nums)

theta = theta_func(2, 100)
# theta = np.linspace(0, 2*np.pi, 100)

x = 2*np.cos(theta) + np.cos(2*theta)
y = 2*np.sin(theta) - np.sin(2*theta)

plt.plot(x, y)
```

```python
theta2 = theta_func(10, 500)

r2 = theta2**2

x2 = r2*np.cos(theta2)
y2 = r2*np.sin(theta2)

plt.plot(x2, y2)
```
