**Problema 3.1 - plotando dados experimentais**

```python
def somat(data, r, i):
  soma = 0
  if i <= r:
    for m in range(-i, r+1):
      soma += data[i+m, 1]
  elif i >= len(data)-r:
    for m in range(-r, len(data)-i):
      soma += data[i+m, 1]
  else:
    for m in range(-r, r+1):
      soma += data[i+m, 1]
  return soma

data = np.loadtxt('sunspots.txt')

r = 5
Yk = np.zeros(len(data[:, 0]))

for i in range(0, len(Yk)):
  Yk[i] += (1/(2*r))*somat(data, r, i)

x = data[:,0]
y = data[:, 1]

plt.xlim(0, 1000)

plt.plot(x, y)
plt.plot(x, Yk)
plt.show()
```
